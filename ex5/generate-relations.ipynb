```python
from math import sin, cos, sqrt, atan2, radians
import csv

# Using Haversine formula of aproximation
def calculate_distance(coord1, coord2):
    # approximate radius of earth in km
    R = 6373.0
    
    lat1 = radians(coord1[0])
    lon1 = radians(coord1[1])
    lat2 = radians(coord2[0])
    lon2 = radians(coord2[1])
    
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance
```

```python
cities = []
with open('cidades.csv', newline='') as csvfile:
    cidades_file = csv.DictReader(csvfile)
    for row in cidades_file:
        cities.append(row)
with open('relations.csv', mode='w+') as relations_file:
    relations_writer = csv.writer(relations_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    relations_writer.writerow(['src', 'dst', 'relationship'])

    for city in cities:
        for compare_city in cities:
            if city['id'] == compare_city['id'] :
                continue;
            else:
                coord1 = (float(city['latitude']), float(city['longitude']))
                coord2 = (float(compare_city['latitude']), float(compare_city['longitude']))
                relations_writer.writerow([city['id'], compare_city['id'], calculate_distance(coord1, coord2)])
                print(city['id'], compare_city['id'])
                print(calculate_distance(coord1, coord2))
```
